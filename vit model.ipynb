```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
from torch.cuda.amp import autocast, GradScaler
from tqdm.auto import tqdm
from transformers import ViTForImageClassification, ViTFeatureExtractor
import pandas as pd
import matplotlib.pyplot as plt
import os

# Set device to GPU (if available)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Data Loading and Preprocessing
IMG_SIZE = (224, 224)  # Resize images to 224x224 for ViT

train_transform = transforms.Compose([
    transforms.Resize(IMG_SIZE),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

valid_transform = transforms.Compose([
    transforms.Resize(IMG_SIZE),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

train_dataset = datasets.ImageFolder('/content/new-plant-diseases-dataset/New Plant Diseases Dataset(Augmented)/New Plant Diseases Dataset(Augmented)/train', transform=train_transform)
valid_dataset = datasets.ImageFolder('/content/new-plant-diseases-dataset/New Plant Diseases Dataset(Augmented)/New Plant Diseases Dataset(Augmented)/valid', transform=valid_transform)

train_dataloader = DataLoader(train_dataset, batch_size=64, shuffle=True, num_workers=4)
valid_dataloader = DataLoader(valid_dataset, batch_size=64, shuffle=False, num_workers=4)

# Load Pretrained Vision Transformer (ViT) model from Hugging Face
model_name = "google/vit-base-patch16-224-in21k"
num_classes = len(train_dataset.classes)
vit_model = ViTForImageClassification.from_pretrained(model_name, num_labels=num_classes)

# Load the feature extractor (for preprocessing images)
feature_extractor = ViTFeatureExtractor.from_pretrained(model_name)
vit_model.to(device)

# Set Optimizer and Loss Function
optimizer = optim.AdamW(vit_model.parameters(), lr=1e-3)
loss_fn = nn.CrossEntropyLoss()

# Mixed Precision Training Setup
scaler = GradScaler()

# Training Step with Mixed Precision
def train_step(model, dataloader, loss_fn, optimizer, device):
    model.train()
    train_loss, train_acc = 0, 0
    train_precision, train_recall, train_f1 = 0, 0, 0

    for X, y in tqdm(dataloader):
        X, y = X.to(device), y.to(device)

        optimizer.zero_grad()

        with autocast():
            y_pred = model(X).logits
            loss = loss_fn(y_pred, y)

        scaler.scale(loss).backward()
        scaler.step(optimizer)
        scaler.update()

        train_loss += loss.item()
        y_pred_class = y_pred.argmax(dim=1)
        train_acc += (y_pred_class == y).sum().item() / len(y)

        precision = (y_pred_class == y).sum().item() / len(y_pred_class)
        recall = (y_pred_class == y).sum().item() / len(y)
        f1 = 2 * (precision * recall) / (precision + recall + 1e-8)

        train_precision += precision
        train_recall += recall
        train_f1 += f1

    train_loss /= len(dataloader)
    train_acc /= len(dataloader)
    train_precision /= len(dataloader)
    train_recall /= len(dataloader)
    train_f1 /= len(dataloader)

    return train_loss, train_acc, train_precision, train_recall, train_f1

# Test Step
def test_step(model, dataloader, loss_fn, device):
    model.eval()
    test_loss, test_acc = 0, 0
    test_precision, test_recall, test_f1 = 0, 0, 0

    with torch.no_grad():
        for X, y in tqdm(dataloader):
            X, y = X.to(device), y.to(device)

            y_pred = model(X).logits
            loss = loss_fn(y_pred, y)

            test_loss += loss.item()
            y_pred_class = y_pred.argmax(dim=1)
            test_acc += (y_pred_class == y).sum().item() / len(y)

            precision = (y_pred_class == y).sum().item() / len(y_pred_class)
            recall = (y_pred_class == y).sum().item() / len(y)
            f1 = 2 * (precision * recall) / (precision + recall + 1e-8)

            test_precision += precision
            test_recall += recall
            test_f1 += f1

    test_loss /= len(dataloader)
    test_acc /= len(dataloader)
    test_precision /= len(dataloader)
    test_recall /= len(dataloader)
    test_f1 /= len(dataloader)

    return test_loss, test_acc, test_precision, test_recall, test_f1

# Save Model
def save_model(model, path):
    torch.save(model.state_dict(), path)
    print(f"Model saved to {path}")

# Save Metrics to CSV
def save_metrics(results, filename):
    df = pd.DataFrame(results)
    df.to_csv(filename, index=False)
    print(f"Metrics saved to {filename}")

# Save Plots
def save_plots(results):
    epochs = range(1, len(results["train_loss"]) + 1)

    plt.figure(figsize=(15, 10))

    plt.subplot(2, 2, 1)
    plt.plot(epochs, results["train_f1"], label="Train F1")
    plt.plot(epochs, results["test_f1"], label="Test F1")
    plt.title("F1 Score")
    plt.xlabel("Epochs")
    plt.ylabel("F1")
    plt.legend()

    plt.subplot(2, 2, 2)
    plt.plot(epochs, results["train_precision"], label="Train Precision")
    plt.plot(epochs, results["test_precision"], label="Test Precision")
    plt.title("Precision")
    plt.xlabel("Epochs")
    plt.ylabel("Precision")
    plt.legend()

    plt.subplot(2, 2, 3)
    plt.plot(epochs, results["train_recall"], label="Train Recall")
    plt.plot(epochs, results["test_recall"], label="Test Recall")
    plt.title("Recall")
    plt.xlabel("Epochs")
    plt.ylabel("Recall")
    plt.legend()

    plt.subplot(2, 2, 4)
    plt.plot(epochs, results["train_acc"], label="Train Accuracy")
    plt.plot(epochs, results["test_acc"], label="Test Accuracy")
    plt.title("Accuracy")
    plt.xlabel("Epochs")
    plt.ylabel("Accuracy")
    plt.legend()

    plt.tight_layout()
    plt.savefig("/content/drive/MyDrive/avit module/training_plots.png")
    print("Plots saved to /content/drive/MyDrive/avit module/training_plots.png")

# Training Loop
def train(model, train_dataloader, valid_dataloader, optimizer, loss_fn, epochs, device):
    results = {
        "train_loss": [], "train_acc": [], "train_precision": [], "train_recall": [], "train_f1": [],
        "test_loss": [], "test_acc": [], "test_precision": [], "test_recall": [], "test_f1": []
    }

    model.to(device)

    for epoch in range(epochs):
        train_loss, train_acc, train_precision, train_recall, train_f1 = train_step(model, train_dataloader, loss_fn, optimizer, device)
        test_loss, test_acc, test_precision, test_recall, test_f1 = test_step(model, valid_dataloader, loss_fn, device)

        print(f"Epoch {epoch+1}/{epochs}")
        print(f"Train - Loss: {train_loss:.4f}, Acc: {train_acc:.4f}, Precision: {train_precision:.4f}, Recall: {train_recall:.4f}, F1: {train_f1:.4f}")
        print(f"Test  - Loss: {test_loss:.4f}, Acc: {test_acc:.4f}, Precision: {test_precision:.4f}, Recall: {test_recall:.4f}, F1: {test_f1:.4f}")

        results["train_loss"].append(train_loss)
        results["train_acc"].append(train_acc)
        results["train_precision"].append(train_precision)
        results["train_recall"].append(train_recall)
        results["train_f1"].append(train_f1)

        results["test_loss"].append(test_loss)
        results["test_acc"].append(test_acc)
        results["test_precision"].append(test_precision)
        results["test_recall"].append(test_recall)
        results["test_f1"].append(test_f1)

    save_metrics(results, "/content/drive/MyDrive/avit module/training_results.csv")
    save_model(model, "/content/drive/MyDrive/avit module/vit_model.pth")
    save_plots(results)

    return results

# Train the model for 10 Epochs
results = train(vit_model, train_dataloader, valid_dataloader, optimizer, loss_fn, epochs=10, device=device)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import torch
from torchvision.utils import make_grid

# Get class labels
class_names = train_dataset.classes  # or valid_dataset.classes

# Function to show images and predictions
def visualize_predictions(model, dataloader, num_images=6):
    model.eval()
    images_so_far = 0

    plt.figure(figsize=(15, 8))

    with torch.no_grad():
        for inputs, labels in dataloader:
            inputs = inputs.to(device)
            labels = labels.to(device)

            outputs = model(inputs).logits
            _, preds = torch.max(outputs, 1)

            for j in range(inputs.size()[0]):
                if images_so_far >= num_images:
                    return

                images_so_far += 1
                img = inputs[j].cpu().numpy().transpose((1, 2, 0))
                img = np.clip(img * np.array([0.229, 0.224, 0.225]) +
                              np.array([0.485, 0.456, 0.406]), 0, 1)

                plt.subplot(num_images // 3 + 1, 3, images_so_far)
                plt.imshow(img)
                plt.title(f"Actual: {class_names[labels[j]]}\nPredicted: {class_names[preds[j]]}")
                plt.axis("off")

        plt.tight_layout()
        plt.show()

# Run it on the validation dataloader
visualize_predictions(vit_model, valid_dataloader, num_images=6)
```

```python
import random
from PIL import Image

# Function to visualize random predictions from across the dataset
def visualize_random_predictions(model, dataset, num_images=6):
    model.eval()
    indices = random.sample(range(len(dataset)), num_images)

    plt.figure(figsize=(15, 8))

    for i, idx in enumerate(indices):
        img, label = dataset[idx]
        input_tensor = img.unsqueeze(0).to(device)

        with torch.no_grad():
            output = model(input_tensor).logits
            pred = output.argmax(dim=1).item()

        # Convert tensor to image
        img_np = img.permute(1, 2, 0).numpy()
        img_np = np.clip(img_np * np.array([0.229, 0.224, 0.225]) +
                         np.array([0.485, 0.456, 0.406]), 0, 1)

        plt.subplot(num_images // 3 + 1, 3, i + 1)
        plt.imshow(img_np)
        plt.title(f"Actual: {class_names[label]}\nPredicted: {class_names[pred]}")
        plt.axis("off")

    plt.tight_layout()
    plt.show()

# Run this on the full validation dataset (random samples)
visualize_random_predictions(vit_model, valid_dataset, num_images=6)
```
